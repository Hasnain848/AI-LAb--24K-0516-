#**Practice Task 1**

```python
class Light:
    def __init__(self, room,status):
        self.status = status
        self.room = room
    def lighton(self):
        self.status = "ON"
        print(f"{self.room}light is turned {self.status}" )
    def switch_off(self):
        self.status = "OFF"
        print(f"{self.room} light is turned {self.status}")
    def check_status(self):
        print(f"{self.room}light is" + self.status)

r2 = Light("Living Room",False)
r2.lighton()
r1 = Light("Bedroom",False)
r1.switch_off()

r2.check_status()
r1.check_status()

```

#**Practice Task 2**

```python
class Staff:
    def __init__(self, name, id_no, dept):
        self.name = name
        self.id_no = id_no
        self.dept = dept

    def display_details(self):
        print(f"Employee Name: {self.name}")
        print(f"ID Number: {self.id_no}")
        print(f"Department: {self.dept}")


class Teacher(Staff):
    def __init__(self, name, id_no, dept, subjects, salary):
        super().__init__(name, id_no, dept)
        self.subjects = subjects
        self.salary = salary

    def show_course_load(self):
        print("Courses Assigned:")
        for i, sub in enumerate(self.subjects, 1):
            print(f"{i}. {sub}")

    def display_details(self):
        super().display_details()
        self.show_course_load()
        print(f"Current Salary: {self.salary}")


class AdminStaff(Staff):
    def __init__(self, name, id_no, dept, role, hours):
        super().__init__(name, id_no, dept)
        self.role = role
        self.hours = hours

    def display_details(self):
        super().display_details()
        print(f"Job Role: {self.role}")
        print(f"Shift Duration: {self.hours}")


class ResearchAssociate(Staff):
    def __init__(self, name, id_no, dept, topic, stipend):
        super().__init__(name, id_no, dept)
        self.topic = topic
        self.stipend = stipend

    def display_details(self):
        super().display_details()
        print(f"Focus Area: {self.topic}")
        print(f"Grant Amount: {self.stipend}")


if __name__ == "__main__":
    t1 = Teacher("Motu Chai wala", 1, "CS", ["DSA", "OOP", "PF"], 1000)
    t1.display_details()
```

#**Practice Task 3**

```python
class BankAccount:
  def __init__(self,account_number,balance):
    self.account_number=account_number
    self.__balance=balance
  def deposit(self,amount):
    if amount < 0:
      print("Invalid deposit amount")
    else:
      self.__balance+=amount
      print(f"Deposited {amount}. New balance: {self.__balance}")
  def withdraw(self,amount):
    if amount <0 or amount > self.__balance:
      print("Invalid withdrawal amount or insufficient balance")
    else:
      self.__balance-=amount
      print(f"Withdrew {amount}. New balance: {self.__balance}")
  def get_balance(self):
    return self.__balance

acc1 = BankAccount(101, 5000)

acc1.deposit(1000)
acc1.withdraw(2000)

print("Current balance:", acc1.get_balance())


print(acc1.__balance)
```
